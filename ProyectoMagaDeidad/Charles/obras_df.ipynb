```python
import pandas as pd
import os
```

```python
obras= pd.read_excel(os.path.join("Data","obras_publicas.xlsx"), header= 3)
obras.head(3)
```

```python
riesgo= pd.read_excel(os.path.join("data","iri_proyectos.xlsx"))
riesgo.head(3)
```

# Procesamiento de las bases de datos

Nos quedamos con las variables más relevantes

```python
list_variables= ["Entidad Pública", "Modalidad de ejecución de la obra", "Tipo de obra - Clasificador Nivel 1","Codigo unico de inversión", "Estado de ejecución","Nivel de gobierno", "Costo de la obra actualizada _2", "Monto Aprobado en soles_2","Departamento", "Sector de la Entidad","Fecha de inicio de obra"]
```

```python
obras1= obras.loc[:,list_variables]
```

```python
date_col=pd.DatetimeIndex(obras1['Fecha de inicio de obra'])
obras1['ano']=date_col.year
```

Identificiando y eliminando observaciones duplicadas y columnas con Nas

```python
obras1.nunique()
```

```python
obras1.drop_duplicates().shape
```

```python
obras2=obras1.drop_duplicates(subset=['Codigo unico de inversión'])
```

```python
obras2.drop(columns=['Fecha de inicio de obra'], inplace=True)
```

```python
obras2.isna().sum()
```

```python
obraswitna=obras2.dropna(subset=['Codigo unico de inversión','Nivel de gobierno','Departamento','ano'])
```

```python
obraswitna.columns
```

```python
list_variables2=["Entidad Pública", "Modalidad de ejecución de la obra", "Tipo de obra - Clasificador Nivel 1","Codigo unico de inversión", "Estado de ejecución","Nivel de gobierno", "Costo de la obra actualizada _2", "Monto Aprobado en soles_2","Departamento", "Sector de la Entidad"]
new_name=['entidad','modalidad_ejecucion','tipo01','cui','estado_ejecucion','nivel_gob','costo_actualizado','monto_aprobado','departamento','sector']
```

```python
obraswitna[new_name]=obraswitna.loc[:,list_variables2]
```

```python
obraswitna.head()
```

```python
obraswitna
```

```python
list_variables4=['ano',
 'entidad',
 'modalidad_ejecucion',
 'tipo01',
 'cui',
 'estado_ejecucion',
 'nivel_gob', 'monto_aprobado',
 'departamento','costo_actualizado']
```

```python
obras_df_new_1=obraswitna.loc[:,list_variables4]
```

```python
obras_df_new_1
```

```python
riesgo_obras=pd.merge(riesgo,
                     obras_df_new_1,
                    left_on = 'cui',
                    right_on = 'cui',
                    how = 'inner', indicator = True)
riesgo_obras
```

```python
list_variables3=["ano","entidad", "modalidad_ejecucion","tipo01","cui","estado_ejecucion","nivel_gob","npry_costoactualizado","monto_aprobado","departamento","npry_puntajeiri"]
```

```python
riesgo_obras2=riesgo_obras.loc[:,list_variables3]
```

```python
riesgo_obras2
```

```python
riesgo_obras2=riesgo_obras2.dropna(subset=["monto_aprobado"])
```

```python
riesgo_obras2['sobrecosto']=riesgo_obras2.eval('(((npry_costoactualizado*100)/monto_aprobado)-100)/100')
```

```python
riesgo_obras2
```

Estableciendo escalas del puntaje de riesgo según lo establecido por la Contraloría

```python
categorias = ['bajo', 'medio', 'alto']
valores = [0, 20, 40, 100]
```

```python
riesgo_obras2['escala_puntaje'] = pd.cut(riesgo_obras2['npry_puntajeiri'],
                              bins = valores, labels = categorias)
```

```python
riesgo_obras2
```

```python
riesgo_obras2.departamento.replace({'CALLAO':'P C DEL CALLAO'}, inplace=True)
```

```python
new_labels=[0,1,2]
new_values=[0,20,40,100]
```

Escala para indice de riesgos

```python
riesgo_obras2['escale'] = pd.cut(riesgo_obras2['npry_puntajeiri'],
                              bins = new_values, labels = new_labels)
```

```python
riesgo_obras2.dropna(subset=["escala_puntaje","escale"],inplace=True)
```

```python
riesgo_obras2['escale']=riesgo_obras2['escale'].astype(str).astype(int)
```

```python
labels_free={0:"0_bajo",1:"1_medio",2:"2_alto"}
riesgo_obras2['escale']=riesgo_obras2.loc[:,'escale'].replace(labels_free)
```
